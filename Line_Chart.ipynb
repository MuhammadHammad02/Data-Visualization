```python
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples)
```

Axis Lables

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')
```

Plotting Multiple Lines

```python
years = range(2000,2010)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.938]
oranges=[0.962,0.942,0.935,0.929,0.920,0.915,0.910,0.908,0.900,0.895]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel("Year")
plt.ylabel("Yield(tons per hactare)")
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.ylabel('Yield (tons per hactare)')
plt.xlabel('Year')

plt.title("Crop Yield in East")
plt.legend(['Apples','Oranges']);
```
