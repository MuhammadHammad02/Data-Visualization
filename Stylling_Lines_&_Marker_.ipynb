```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = range(2000,2010)
apples=[0.895,0.91,0.919,0.926,0.929,0.931,0.934,0.936,0.937,0.938]
oranges=[0.962,0.942,0.935,0.929,0.920,0.915,0.910,0.908,0.900,0.895]
```

```python
plt.plot(years, apples, marker = "o")
plt.plot(years, oranges, marker = "x")
plt.ylabel('Yield (tons per hactare)')
plt.xlabel('Year')

plt.title("Crop Yield in East")
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years, apples, "s-b")
plt.plot(years, oranges, "o-r")
plt.ylabel('Yield (tons per hactare)')
plt.xlabel('Year')

plt.title("Crop Yield in East")
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(years, oranges, 'or')
plt.title('Yield of Oranges(tons per hactare)')
```
